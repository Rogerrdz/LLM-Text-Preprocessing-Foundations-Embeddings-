
### List available models
Let's see what models are available with your API key:

```python
from google import genai

client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])
print("Modelos disponibles:")
for model in client.models.list():
    print(f"  - {model.name}")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

First Message :

```python
from langchain_core.messages import HumanMessage

response = model.invoke([HumanMessage(content="Hola, ¿qué es LangChain?")])
print(response.content)
```

Prompt Template + chain

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres un asistente útil."),
    ("user", "{input}")
])

chain = prompt | model | StrOutputParser()

response = chain.invoke({"input": "¿Qué es un LLM?"})
print(response)
```

### Translation Chain Example
Using LangChain to translate text between languages:

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Prompt template for traduction
prompt = ChatPromptTemplate.from_messages([
    ("system", "Translate the following text to {language}:"),
    ("user", "{text}")
])

chain = prompt | model | StrOutputParser()

# English
result1 = chain.invoke({
    "language": "English",
    "text": "Hola, ¿cómo estás? Me llamo roger y estoy aprendiendo LangChain."
})
print("Translated to English:")
print(result1)
print()

# French
result2 = chain.invoke({
    "language": "French", 
    "text": "Hello, I am learning about artificial intelligence."
})
print("Translated to French:")
print(result2)
```
